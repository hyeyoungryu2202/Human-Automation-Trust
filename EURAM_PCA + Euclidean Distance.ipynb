```python
import pandas as pd
# load dataset into Pandas DataFrame
df = pd.read_excel('trust_cluster_pca.xlsx', names=['trust_clusters', 'transparency', 'inter_personality', 'validity', 'privacy', 'purpose', 'previous_experience'])
df
```
